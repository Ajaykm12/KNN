```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from scipy.io.arff import loadarff
from sklearn.model_selection import cross_val_score
```

```python
# Selection of the Data and converting data into numpy format for flexibility in cleaning
#phising_data_array = np.array(phishing_data_raw[0])
#phising_data_array

phishing_Data_array=np.array(phishing_data_raw[0])
phishing_Data_array
```

```python
# Converting the numpy array into Pandas data frame  and casting the coloumns to numeric type
#phising_data_frame = pd.DataFrame(phising_data_array).apply(pd.to_numeric)
#phising_data_frame[0:10]

phising_data_frame= pd.DataFrame(phishing_Data_array)
phising_data_frame.head(10)
```

```python
#now convert the data frame to numeric

phising_data_frame_num=phising_data_frame.apply(pd.to_numeric)
phising_data_frame_num.head(10)
```

```python
#Duplicate
#col_names = list(phising_data_frame)# Getting Column names of the pandas data frame
#print("Coloumn_names:" + ", ".join(col_names))
# The scikit learn package takes data in form of labels and predictors
# Extracting Result coloum to pass as target coloum
#labels = phising_data_frame["Result"].values
#print(labels)
```

```python
col_names=list(phising_data_frame_num) # will display all column names
col_names
print("column names:"+",".join(col_names)) # joining all the column names
# The scikit learn package takes data in form of labels and predictors
# Extracting Result coloum to pass as target coloum
labels=phising_data_frame_num["Result"].values
print(labels)
len(col_names)
```

```python
#selecting the predictor column names
predictor_cols=col_names[0:len(col_names)-2] # excluding last 2 columns
print("predictor_cols"+",".join(predictor_cols))
#extracting predictor points
predictors=phising_data_frame_num[predictor_cols].values
predictors

```

```python
# Defining The parameters in KNN Classifier
# here we are taking k (neighbour value as 10). 
# its better to take k values from 5 to 20 and do a cross validation and get the best score.
#clf_knn = KNeighborsClassifier(
    #n_neighbors=10,
    #weights='distance'
    #)
#clf_knn = clf_knn.fit(predictors,labels)
```

```python
# put a for loop here and give k=5-20 in for loop and inside for loop give cv=5 or cv=10 and chk diff avg cv values.
#whichever has highest mean will be the k value
```

```python
for k in range(5,20):
    clf_knn = KNeighborsClassifier(
    n_neighbors=k,
    weights='distance'
    )
    clf_knn = clf_knn.fit(predictors,labels)
    score_knn = cross_val_score(clf_knn, predictors, labels, cv=5)
    print("Cross Validation score : " + str(score_knn))
    print("Cross Validation Mean score : " + str(score_knn.mean()))

    
```

```python
# so when you are doing cv=5, we get max avg for k-13 i.e 0.955938976913
# lets chk for cv=10.
```

```python
for k in range(5,20):
    clf_knn = KNeighborsClassifier(
    n_neighbors=k,
    weights='distance'
    )
    clf_knn = clf_knn.fit(predictors,labels)
    score_knn = cross_val_score(clf_knn, predictors, labels, cv=10)
    print("Cross Validation score : " + str(score_knn))
    print("Cross Validation Mean score : " + str(score_knn.mean()))
```

```python
# now we do cv=10, an chk the k value
for k in range(5,20):
    clf_knn = KNeighborsClassifier(
    n_neighbors=k,
    weights='distance'
    )
    clf_knn = clf_knn.fit(predictors,labels)
    score_knn = cross_val_score(clf_knn, predictors, labels, cv=10)
    print("Cross Validation score : " + str(score_knn))
    print("Cross Validation Mean score : " + str(score_knn.mean()))
```
